## Dev에서 가장 f1 score가 높은 Epoch에서의 Test score

```python
import re
data_path="log_50.log"

f = open(data_path, 'r')
flag=0
flag_max=0
flag_max_dev=0
max_f1_score=0
max_f1_score_dev=0
epoch=0
m_list=[]
m_list_dev=[]
#regex = re.compile("\d\d/\d\d/\d\d\d")
while True:
    line = f.readline()
    if not line: break
    check=line[25:32]
    #print(len(check))
    if check == "disease":
        flag = 1
    if len(check) != 0 :
        if check[0] == "[":
            flag=0
        if line[20] == "[":
            flag=0
        #print(check[20])
    if flag == 1:
        #print(line[-13:-10])
        if line[-14:-10]=="Test":
            flag_max=1
        if line[-13:-10]=="Dev":
            flag_max_dev=1
        if flag_max_dev==1 and line[:5]=="macro":
            flag_max_dev=0
            m_list_dev.append(float(line[34:40]))
        if flag_max==1 and line[:5]=="macro":
            flag_max=0
            m_list.append(line)
dev_max_f1=max(m_list_dev)
dev_max_f1_epoch=m_list_dev.index(max(m_list_dev))

print("dev에서 max f1 score Epoch: ",dev_max_f1_epoch)
print("dev에서 max f1 score: ",dev_max_f1)
print()
print("dev에서 max f1 score였던 Epoch에서 test matrix:")
print()
print("            precision    recall  f1-score   support")
print(m_list[dev_max_f1_epoch])          
#print("max_f1_score Epoch: ",m_list.index(max(m_list)))
#print("max_f1_score : ",max(m_list))
```
