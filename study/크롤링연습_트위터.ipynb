```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
from selenium.webdriver.common.desired_capabilities import  DesiredCapabilities
import time
from selenium.webdriver.common.keys import Keys
import datetime as dt
import pandas as pd
```

```python
driver = webdriver.Chrome('C:/Users/SAMSUNG/Downloads/chromedriver_win32 (1)/chromedriver.exe')
```

```python
startdate=dt.date(year=2019,month=11,day=27) #시작날짜
untildate=dt.date(year=2019,month=11,day=28) # 시작날짜 +1
enddate=dt.date(year=2019,month=11,day=29) # 끝날짜
```

```python
#https://twitter.com/search?q=%EA%B9%80%EC%84%9C%ED%98%95%20since%3A2019-01-28%20until%3A2019-01-29&amp;amp;amp;amp;amp;amp;lang=eg
query="동백이"
totaltweets=[] #검색된 트윗들을 담을 리스트 형성

while not enddate==startdate:
    url='https://twitter.com/search?q='+query+'%20since%3A'+str(startdate)+'%20until%3A'+str(untildate)+'&amp;amp;amp;amp;amp;amp;lang=eg'
    driver.get(url)
    html = driver.page_source #그 주소의 html을 가지고 오는 함수
    soup=BeautifulSoup(html,'html.parser', from_encoding='utf-8')
    
    lastHeight = driver.execute_script("return document.body.scrollHeight") #화면 높이만큼 스크롤하기위해
    
    #본격적인 자동 건색과 크롤링을 위한 while 구문
    while True:
        
        dailyfreq={'Date':startdate}

        wordfreq=0
        tweets=soup.find_all("p", {"class": "TweetTextSize"})
                    
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)
        
        #새롭게 로드된 페이지가 newHeight가 된다.
        newHeight = driver.execute_script("return document.body.scrollHeight")
        
        if newHeight != lastHeight:
            html = driver.page_source
            soup=BeautifulSoup(html,'html.parser', from_encoding='utf-8')
            tweets=soup.find_all("p", {"class": "TweetTextSize"})
            wordfreq=len(tweets)
        else: #다 긁어왔을때 새로운 것을 넣기 위해 초기화
            dailyfreq['Frequency']=wordfreq
            wordfreq=0
            startdate=untildate
            untildate+=dt.timedelta(days=1)
            dailyfreq={}
            totaltweets.append(tweets)
            break

        lastHeight = newHeight
```

```python
df = pd.DataFrame(columns=['id','message'])
```

```python
number=1
for i in range(len(totaltweets)):
    for j in range(len(totaltweets[i])):
        df = df.append({'id': number,'message':(totaltweets[i][j]).text}, ignore_index=True)
        number = number+1
```
