```python
import pandas as pd

read_file = pd.read_excel (r"데이터.xlsx") 
read_file.to_csv (r'new_data.csv', index = None, header=True)

read_file = pd.read_excel (r"출석여부파악.xlsx")
read_file.to_csv (r'new_출석여부파악.csv', index = None, header=True)
```

```python
data=pd.read_csv("new_data.csv") #user가 추가적으로 생성하는 df --> 선생님들이 추가로 입력한다는 file!!
data.head()
```

```python
data_filter=data[['학년','반','번호_','수강자','강의명','수강 시작 일시','수강 종료 일시','상태']] #필요없는 column remove
data_filter
```

```python
data_want=pd.read_csv("new_출석여부파악.csv") #data.xlsc
data_want.head()
```

```python
data_want_re=data_want[['학년','반','번호','성명','강의명','수강시작일시','수강종료일시','상태']]
data_want_re.head()
```

```python
data_want_filter=data_want[['학년','반','번호']]
data_want_filter.head()
```

```python
final_pd=[] #최종적으로 data_want_filter df에 붙일 df
for i in range(len(data_want_filter)): #data_want_filter에서 한 row 씩 읽어서 비교
    flag=0
    temp_pd=[] #한 row가 되고, final_pd에 append시켜줄 애
    want=data_want_filter.loc[[i]].values.tolist()
    for j in range(len(data_filter)):
        want_find=data_filter.loc[[j]].values.tolist()
        want_find_=want_find[0][:3]
       # print(want[0])
       # print(want_find_)
       # print("+++++++++++++++++++++++++++")
        if want[0]==want_find_:
            flag=1
            #print(want_find)
            temp_pd=want_find[0]
            final_pd.append(temp_pd)
    if flag==0:
        final_pd.append(data_want_re.loc[[i]].values.tolist()[0])
#print(final_pd)
final_df=pd.DataFrame.from_records(final_pd,columns=['학년','반','번호','수강자','강의명','수강 시작 일시','수강 종료 일시','상태'])       
```

```python
import numpy as np
final_df['출결여부']=np.where(final_df['상태'] == '완강', '출석','결석')
```
