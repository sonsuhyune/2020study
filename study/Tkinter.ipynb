## canvas에 line 그리기




```python
from tkinter import *

root=Tk()

canvas=Canvas(root, width=200, height=150, bg="white",bd=2)
canvas.pack()

canvas.create_line(0,75,100,75,fill="blue")

root.mainloop()
```

## canvas에 line 그리기


```python
from tkinter import *

root=Tk()

canvas=Canvas(root, width=200, height=150, bg="white",bd=2)
canvas.pack()

for i in range(50,300,50):
    canvas.create_line(i,0,i,200,fill="blue")

for j in range(50,300,50):
    canvas.create_line(0,j,300,j,fill="blue")

root.mainloop()
```

## 가로가 80, 세로가 80인 사각형을 정중앙에



```python
from tkinter import *

root=Tk()

canvas=Canvas(root, width=200, height=150, bg="white",bd=2)
canvas.pack()

canvas.create_rectangle(60,35,140,115,fill='red')

root.mainloop()
```

## 반지름이 50인 원을 정중앙에


```python
from tkinter import *

root=Tk()

canvas=Canvas(root, width=200, height=150, bg="white",bd=2)
canvas.pack()

canvas.create_oval(50,25,150,125,fill='red')

root.mainloop()
```
