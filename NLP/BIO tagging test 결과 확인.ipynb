```python
import csv
full_text=[]
with open("NER_result_conll.txt", newline = '') as games:                                                                                          
    game_reader = csv.reader(games, delimiter='\t')
    for game in game_reader:
        full_text.append(game)
        

```

```python
full_text #1번 index: 모델이 예측한 것
          
```

```python
flag=0
for j in full_text:
    if j==[]:
        continue
        
    i=j[0].split()
    if i[0]==".":
        print(i[0])
        continue
    if i[1]=="O-MISC": #O라고 예측하면
        if flag==0:
            print(i[0], end=" ")
        else:
            print("\"[desease]",end=" ")
            print(i[0], end=" ")
            flag=0
    elif i[1]=="B-MISC":
        print("\"", end="")
        print(i[0], end=" ")
        
    elif i[1]=="I-MISC":
        flag=1
        print(i[0], end=" ")
        
```
